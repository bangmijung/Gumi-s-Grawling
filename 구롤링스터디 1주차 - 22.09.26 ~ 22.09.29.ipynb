```python
#BS4 test (ÏÑúÍ∞ïÎåÄÌïôÍµê ÏúÑÌÇ§ÌîºÎîîÏïÑ)

import requests
from bs4 import BeautifulSoup

webpage = requests.get("https://en.wikipedia.org/wiki/Sogang_University")
soup = BeautifulSoup(webpage.content, "html.parser")
print(soup)
```

```python
# ÌÉúÍ∑∏ ÌÉêÏÉâÌïòÍ∏∞ (span)
# find all -> Î¶¨Ïä§Ìä∏ Î∞òÌôò
soup.find_all('span')
```

```python
# ÌÉúÍ∑∏ ÏÜçÏÑ±Îì§ÏùÄ ÎπºÍ≥† Í∑∏ ÏïàÏóê ÏûàÎäî ÌÖçÏä§Ìä∏Îßå ÍπîÎÅîÌïòÍ≤å Í∞ÄÏ†∏Ïò§Í≥† Ïã∂Îã§Î©¥?
# soup.find_all('span')[0].string
# soup.find_all('span')[0].text

soup.find_all('span')[0].string
```

```python
# html ÏÜçÏÑ± ÌôúÏö©ÌïòÏó¨ Ï∞æÍ∏∞
# attrs ÌååÎùºÎØ∏ÌÑ∞Î•º ÏßÄÏ†ïÌïòÎ©¥ ÎêúÎã§. ÎîïÏÖîÎÑàÎ¶¨ ÌòïÌÉúÎ°ú ÎÑ£ÏúºÎ©¥ Îê®(Ïó¨Îü¨Í∞ú ÏßÄÏ†ï Í∞ÄÎä•)
# ÎßåÏïΩ mw-page-title-mainÏù¥ÎùºÎäî Ïù¥Î¶ÑÏùò classÎßå Í∏ÅÏñ¥Ïò§Í≥† Ïã∂Îã§Î©¥?

soup.find_all(attrs={'class':'mw-page-title-main'})[0].string
```
